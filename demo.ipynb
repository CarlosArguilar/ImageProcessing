## Custom filters

### Gaussian Blur 5x5

```python
custom_filter = (1/256)*np.array([
    [1,4,6,4,1],
    [4,16,24,16,4],
    [6,24,36,24,6],
    [4,16,24,16,4],
    [1,4,6,4,1],
])
```

```python
res = imp.apply_filter(custom_filter)
```

### Sharpen

```python
custom_filter = [
    [0,-1,0],
    [-1,5,-1],
    [0,-1,0]
]
```

```python
res = imp.apply_filter(custom_filter)
```

### Ridge

```python
custom_filter = [
    [-1,-1,-1],
    [-1,8,-1],
    [-1,-1,-1],
]
```

```python
res = imp.apply_filter(custom_filter)
```
